```python
train_arr = train_arr / 255
```

```python
y_labels = np.zeros(labels_arr.shape)
for i in range(8000):
    if labels_arr[i] == 2:
        y_labels[i] = 1
    if labels_arr[i] == 3:
        y_labels[i] = 2
    if labels_arr[i] == 6:
        y_labels[i] == 3
```

```python
classes = [ '0', '2' , '3', '6']
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(train_arr, y_labels , test_size=0.3)
```

```python
from tensorflow import keras 
```

```python
model = keras.Sequential([
    keras.layers.Dense(256,activation='relu',input_shape=(784,)),
    keras.layers.Dense(256, activation='relu'),
    keras.layers.Dense(256, activation='relu'),
    keras.layers.Dense(4, activation='softmax') ])
```

```python
model.compile(optimizer='adam', 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'] )
```

```python
model.fit(train_x, train_y, batch_size=10 , epochs = 33)
```

```python
accuracy = model.evaluate(test_x,test_y,batch_size=10)
print(accuracy[1])
```

```python
model.save('cv_model.h5')
```

```python
predictions = model.predict_classes(test_arr, batch_size=10)
```
